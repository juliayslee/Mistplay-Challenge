### 1.  General Visualisation

Using the takeHome.csv data, plot differences in LTV (life time value), ttime (total time) and tue (total units earned) across the different ab groups.

```python
# import all libraries and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('takeHome.csv')
```

```python
df.head(20)
```

```python
print(df.shape)
print(df.dtypes)
print(df.info())
```

```python
# ltv has 1 nulls, tue has 2 nulls
df[pd.isnull(df['ltv'])], df[pd.isnull(df['tue'])]
```

### Clean dataset 

1. remove dictionary format for data values
2. drop missing values 
3. convert data types into numeric/int

```python
# remove {"n": value} format 
df = df.replace([r'{"n":"', r'"}'], '', regex=True)
# drop 3 rows with missing values
df = df.dropna()
# convert to numeric
df = df.apply(pd.to_numeric)
# check 
print(df.info())
df.head()
```

```python
df.abt.nunique()
```

```python
print('Data Information:', '\n', df.describe())
```

ltv, ttime and tue columns all have many outliers well beyond the 75th percentile.

```python
# plot distribution plots for each column individually
f, axes = plt.subplots(2,2, figsize=[8,6])
sns.set(style='white', palette='pastel')
sns.despine()

sns.distplot(df['ltv'], color='green', ax=axes[0,0])
sns.distplot(df['ttime'], color='red', ax=axes[0,1])
sns.distplot(df['tue'], color='blue', ax=axes[1,0])
sns.distplot(df['abt'], color='orange', ax=axes[1,1])

plt.tight_layout();
```

The graphs above show the extreme skewed distribution in ltv, ttime and tue, revealing again the outliers within each column. Abt is data for the AB testing groups, which was not necessary to graph but is interesting to see the distribution of users within each group. Next we'll take a look at the differences in each column across the AB testing groups.

```python
# there are 60 unique AB groups

g = sns.FacetGrid(df, col='abt', col_wrap=6, height=1.8, xlim=(0,5000)) #xlim set to reduce skew
g.map(sns.distplot, 'ltv')
plt.subplots_adjust(top=0.95)
g.fig.suptitle('Life Time Value')

g = sns.FacetGrid(df, col='abt', col_wrap=6, height=1.8, xlim=(0,1.25e8))
g.map(sns.distplot, 'ttime')
plt.subplots_adjust(top=0.95)
g.fig.suptitle('Total Time')

g = sns.FacetGrid(df, col='abt', col_wrap=6, height=1.8, xlim=(0,2000))
g.map(sns.distplot, 'tue')
plt.subplots_adjust(top=0.95)
g.fig.suptitle('Total Units Earned');
```

The distributions of each column, Life Time Value, Total Time and Total Units Earned do not show any alarming trends. The distributions are consistent across each AB group.

### 2. Game Economy

Using takeHome.csv, can you provide insight on how the time spent and units earned should relate to GXP and PXP, to further optimize the Mistplay economy by maximizing the LTV of users, while minimizing the total units earned (tue)?

Mistplay Mechanism:
1. Users accumulate GXP (Game Experience) by spending time on a game
2. Users level up in game -> earn PXP (Player Experience) points
3. Users level up in Mistplay when PXP threshold reached

Optimizing the Mistplay economy by maximizing LTV while minimizing tue would require insight on how GXP and PXP affect those markers. Let's first take a look at how users in each "economy" behave.

- Control [1-6], [55-60]
- New Economy 1 [7-14], [19-34]
- New Economy 2 [15,16], [35 - 44]
- New Economy 3 [17,18], [45 - 54]

```python
# define each abt group with the group range and replace with economy type
control = list(range(1,7)) + list(range(55,61))
new_econ1 = list(range(7,15)) + list(range(19,35))
new_econ2 = list(range(15,17)) + list(range(35,45))
new_econ3 = list(range(17,19)) + list(range(45,55))

df['abt'] = df['abt'].replace(control, 'control').replace(new_econ1, 'New Economy 1').replace(new_econ2, 'New Economy 2').replace(new_econ3, 'New Economy 3')
```

```python
# graph lvt, ttime, tue for each economy
palette = sns.color_palette('pastel')

plt.figure(figsize = (10,20))
plt.subplot(3, 1, 1)
sns.violinplot(data=df, x='abt', y='ltv', palette=palette)
plt.title('Life Time Value of Different Economies', size=14)
plt.xlabel('Economy Groups')
plt.ylabel('Life Time Value')

plt.figure(figsize = (10,20))
plt.subplot(3, 1, 1)
sns.violinplot(data=df, x='abt', y='ttime', palette=palette)
plt.title('Total Time Spent Across Different Economies', size=14)
plt.xlabel('Economy Groups')
plt.ylabel('Total Time')

plt.figure(figsize = (10,20))
plt.subplot(3, 1, 1)
sns.violinplot(data=df, x='abt', y='tue', palette=palette)
plt.title('Total Units Earned for Different Economies', size=14)
plt.xlabel('Economy Groups')
plt.ylabel('Total Units Earned')

plt.tight_layout();
```

```python
print(df[df['ltv'] == df['ltv'].max()])
print(df[df['tue'] == df['tue'].max()])
print(df[df['ttime'] == df['ttime'].max()])
```

There are large outliers across all economies individually. New Economy 2 seems to have the highest range in Life Time Value, including the highest Life Time Value value. Although highest LTV, this particular user does not have the most tue or the most ttime. Next we will take a look at the relationship between ltv vs. tue, ttime vs. ltv, and ttime vs. tue across all economies. From what we know about the Mistplay mechanism, I will assume a linear model for the following plots.

```python
sns.lmplot(x='ltv', y='tue', hue='abt', data=df, palette='Set1', height=4, aspect=2)
plt.title('Total Units Earned over Life Time Value', size=14)
plt.xlabel('Life Time Value')
plt.ylabel('Total Units Earned')

sns.lmplot(x='ttime', y='ltv', hue='abt', data=df, palette='Set1', height=4, aspect=2)
plt.title('Life Time Value over Total Time', size=14)
plt.xlabel('Total Time')
plt.ylabel('Life Time Value')

sns.lmplot(x='ttime', y='tue', hue='abt', data=df, palette='Set1', height=4, aspect=2)
plt.title('Total Units Earned over Total Time', size=14)
plt.xlabel('Total Time')
plt.ylabel('Total Units Earned');
```

The first plot represents the TUE over the LTV. New Economy 3 shows the most favourable situation in obtaining a high Life Time Value while minimizing Total Units Earned. New Economy 3 also spends the longest Total Time without earning the most Total Units. 
As New Economy 3 represents Mistplay's optimal outcome, let's now take a look at how GXP and PXP relate to the LTV and TUE.

```python
# read in econ.txt file and explore dataset
econ = pd.read_csv('econs.txt', sep='\t')

econ.head(40)
econ.shape
econ.describe()
econ.info()
```

Let's take a look at the 'econ.txt' dataset in its entirety. This dataset describes previous economies that were tested.

```python
corr = econ.corr()

plt.figure(figsize = (8,8))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True      #masking the upper half of triangle to cut redundancy
sns.heatmap(corr, annot = True, fmt = '0.2f', cmap = 'vlag_r', 
           center = 0, square = True, mask = mask)
plt.ylim(len(corr), 0.5);
```

Correlation map above is a visual representation of Mistplay's game mechanism stated previously.

```python
print('\033[1m Sum of Units per Economy \033[0m', '\n', econ.groupby('econ')['cumUnits'].sum())
print('\n')
print('\033[1m Sum of Total Time Spent per Economy \033[0m', '\n', econ.groupby('econ')['cumTime'].sum())
print('\n')
print('\033[1m Sum of Level Up per Economy \033[0m', '\n', econ.groupby('econ')['lvl'].sum())
print('\n')
print('\033[1m Sum of GXP per Economy \033[0m', '\n', econ.groupby('econ')['cumGXP'].sum())
print('\n')
print('\033[1m Sum of PXP per Economy \033[0m', '\n', econ.groupby('econ')['cumPXP'].sum())
```

Before diving into some visuals, some questions that are driving this portion of the analysis are:
- What is the relationship between GXP and PXP? 
- How does GXP and PXP relate to the Total Units Earned (cumUnits)?
- How long does it take to earn units?
- How long does it take to gain GXP?
- How long does it take to gain PXP (level up)?

Answering these questions will give us insight on how the time spent and units earned relate to GXP and PXP.

```python
# Graphs to answer the above questions
plt.figure(figsize = (10,20))
plt.subplot(6, 1, 1)
sns.lineplot(x='cumGXP', y='cumPXP', hue='econ', data=econ, palette='rainbow')
plt.legend(loc='lower right')
plt.title('How does GXP and PXP relate to each other?', size=14)

plt.figure(figsize = (10,20))
plt.subplot(6, 1, 2)
sns.lineplot(x='cumGXP', y='cumUnits', hue='econ', data=econ, palette='rainbow')
plt.title('How does GXP relate to Units Earned?', size=14)

plt.figure(figsize = (10,20))
plt.subplot(6, 1, 3)
sns.lineplot(x='cumPXP', y='cumUnits', hue='econ', data=econ, palette='rainbow')
plt.legend(loc='lower right')
plt.title('How does PXP relate to Units Earned?', size=14)

plt.figure(figsize = (10,20))
plt.subplot(6, 1, 4)
sns.lineplot(x='cumTime', y='cumUnits', hue='econ', data=econ, palette='rainbow')
plt.legend(loc='lower right')
plt.title('How long does it take to earn Units?', size=14)

plt.figure(figsize = (10,20))
plt.subplot(6, 1, 5)
sns.lineplot(x='cumTime', y='cumGXP', hue='econ', data=econ, palette='rainbow')
plt.legend(loc='lower right')
plt.title('How long does it take to gain GXP?', size=14)

plt.figure(figsize = (10,20))
plt.subplot(6, 1, 6)
sns.lineplot(x='cumTime', y='lvl', hue='econ', data=econ, palette='rainbow')
plt.legend(loc='lower right')
plt.title('How long does it take to Level Up?', size=14)

plt.tight_layout();
```

Generally speaking, and increase in GXP results in an increase in PXP as well. Economy 3 gains the most PXP given the least GXP increase, which means that the users in this economy are leveling up in games faster than the other economies. This is also the case in our last graph where Economy 3 levels up faster than the rest as well. Although Economy 3 is leveling up the fastest, they are not earning the most units, whereas Economy 1 is spending the least amount of time, but earning the most units the fastest. This result can mean perhaps that users in Economy 3 are motivated by "leveling up" in a game/ on the Mistplay platform over earning units. Users here seem to spend time focusing on increasing levels than anything else. 

### Conclusion
To optimize Mistplay's economy by maximizing Life Time Value while minimizing the Total Units Earned, I would advise further investigation on the "leveling up" aspect of game play. Heightening this experience may lead to more motivation to continue playing the game and as a result an increase in Life Time Value without sacrificing units given.
